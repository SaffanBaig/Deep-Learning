```python
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.fashion_mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
#Normalize
train_images = train_images/255.0
test_images = test_images/255.0
```

```python
#1-Flatten = Convert n dimensions in to 1 dimension
#2- relu = If value greater than zero (accept) else for zero and negative convert into zero
#3- softmax = Use when we have multiple classification (e.g Trouser, Pants etc)
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(),
             loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
#Train
model.fit(train_images, train_labels, epochs=5, batch_size=128)
```

```python
model.evaluate(test_images, test_labels)
```
